```python
#import important libraries
import pandas as pd
import numpy as np
import matplotlib as mtb
import seaborn as sns
```

```python
#read the dataset
file_path= r"C:\Users\91835\AB_NYC_2019.csv"
abc=pd.read_csv(file_path)
abc
```

```python
#find count of missing values
missing_values=abc.isnull().sum()
print(missing_values)
```

```python
#drop all null values
abc_cleaned=abc.dropna()
```

```python
#convert argument datatype to the datatype of column
abc_cleaned['price']=pd.to_numeric(abc_cleaned['price'])
abc_cleaned['last_review']=pd.to_datetime(abc_cleaned['last_review'])
abc_cleaned['reviewss_per_month']=pd.to_numeric(abc_cleaned['reviews_per_month'])
abc_cleaned['minimum_nights']=pd.to_numeric(abc_cleaned['minimum_nights'])
abc_cleaned['calculated_host_listings_count']=pd.to_numeric(abc_cleaned['calculated_host_listings_count'])
abc_cleaned['availability_365']=pd.to_numeric(abc_cleaned['availability_365'])
abc_cleaned['name']=abc_cleaned['name'].astype(str)
```

```python
#cleaned dataset with no null values
abc_cleaned
```

```python
#forward fill method to fill missing values
abc_cleaned['name'].fillna(method='ffill', inplace=True)
```

```python
#finding missing values in cleaned dataset
missing_values=abc_cleaned.isnull().sum()
print(missing_values)
```

```python
#drop all duplicates from the dataset
duplicates1=abc_cleaned.drop_duplicates()
```
